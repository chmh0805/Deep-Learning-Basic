```python
import os
now_dir = os.getcwd()
txt_name = now_dir + "\\parking_stat.txt"
parking_list = [] # 주차한 차 리스트
parking_area = 20 # 주차장 자리 갯수

menu = '''
================
    1. 입차
    2. 출차
    3. 현황
    4. 초기화
    5. 종료
================
'''

class Car:
    def __init__(self, car_number):
        self.car_number = car_number
    
    def __str__(self):
        return '''
        차번호 : {}
        '''.format(self.car_number)


class Park(Car):    
    def park(self):
        f1 = open(txt_name, 'a')
        f2 = open(txt_name, 'r')
        if len(f2.read().split()) == parking_area:
            print("주차장이 만석입니다.")
        else:
            parking_list.append(self.car_number)
            f1.write("{}\n".format(self.car_number))
            print("주차 등록이 완료되었습니다.")
        f1.close()
        f2.close()


def park_out():
    f3 = open(txt_name, 'r')
    while True:
        print(" 현재 주차 목록 ")
        print("================")
        for car in parking_list:
            print(car)
        print("================")
        out_input = input("\n출차하는 차량 번호를 입력하세요.")
        if out_input not in parking_list:
            print("차량번호를 확인해주세요.")
            continue
        else:
            parking_list.pop(parking_list.index(out_input))
            print("출차 완료되었습니다.")
            with open(txt_name, 'w') as ff:
                for car in parking_list:
                    ff.write("{}\n".format(car))


def now_stat():
    with open(txt_name, 'r') as f1:
        now_list = f1.read().split()
    print('')
    print("주차 현황 : {} / {}".format(len(now_list), parking_area))
    print(now_list)


def dayopen():
    with open(txt_name, 'w') as f1:
        pass


def user_choice():
    while True:
        user_choice = input(menu)
        try:
            int(user_choice)
        except:
            print("1, 2, 3, 4, 5 중 입력하세요.")
            continue
        if user_choice not in "12345":
            print("1, 2, 3, 4, 5 중 입력하세요.")
            continue
        else:
            break
    return user_choice


def more():
    while True:
        choice = input("추가로 입력하시겠습니까?(y or n)").lower()
        if choice not in ['y', 'n']:
            print("y 또는 n을 입력하세요.")
            continue
        elif choice == 'y':
            return True
        else:
            return False
```

```python
while True:
    user_menu = user_choice()
    
    if user_menu == '5':
        print("시스템을 종료합니다.")
        break
    
    if user_menu == '3':
        now_stat()
    
    if user_menu == '1':
        user_number = input("차번호를 입력하세요. : ")
        user_car = Park(user_number)
        user_car.park()
    
    if user_menu == '2':
        park_out()
        break
    
    if user_menu == '4':
        while True:
            choice1 = input("정말 초기화하시겠습니까?(y or n)").lower()
            if choice1 not in ['y', 'n']:
                print("y 또는 n을 입력해주세요.")
                continue
            elif choice1 == 'y':
                dayopen()
                break
            else:
                break

    if more():
        continue
    else:
        print("시스템을 종료합니다.")
        break
```
