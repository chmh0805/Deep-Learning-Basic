# XOR 게이트

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

### GPU 리소스 절약

```python
from tensorflow.compat.v1 import ConfigProto,InteractiveSession
config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction=0.4
session = InteractiveSession(config=config)
```

#### 데이터 수집

```python
x_data = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

y_xor_data = [
    [0],
    [1],
    [1],
    [0]
]
```

#### 데이터 가공

```python
x_data = np.array(x_data)
y_xor_data = np.array(y_xor_data)
```

#### 모델 선정

```python
xor_model = keras.Sequential()
```

```python
xor_model.add(layers.Dense(4, activation='sigmoid', input_shape=(2,)))
xor_model.add(layers.Dense(2, activation='sigmoid'))
xor_model.add(layers.Dense(1, activation='sigmoid'))
```

```python
optimizer = keras.optimizers.SGD(lr=0.1)
xor_model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

#### 모델 학습

```python
xor_model.fit(x_data, y_xor_data, batch_size=4, epochs=8000)
```

#### 모델 평가

```python
answer = xor_model.predict(x_data)
```

```python
print(np.round(answer))
```

# 실습

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.compat.v1 import ConfigProto,InteractiveSession
config=ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction=0.4
session=InteractiveSession(config=config)
```

```python
# Inputs and Labels
data = [
    [0.72,0.82,-1],
    [0.91,-0.69,-1],
    [0.03,0.93,-1],
    [0.12,0.25,-1],
    [0.96,0.47,-1],
    [0.8,-0.75,-1],
    [0.46,0.98,-1],
    [0.66,0.24,-1],
    [0.72,-0.15,-1],
    [0.35,0.01,-1],
    [-0.11,0.1,1],
    [0.31,-0.96,1],
    [0.0,-0.26,1],
    [-0.43,-0.65,1],
    [0.57,-0.97,1],
    [-0.72,-0.64,1],
    [-0.25,-0.43,1],
    [-0.12,-0.9,1],
    [-0.58,0.62,1],
    [-0.77,-0.76,1]
]
```

```python
data = np.array(data)
```

```python
x_data = data[:,:2]
y_data = (data[:,2] + 1)/2
```

```python
#plt.scatter(data[:,0],data[:,1],c=data[:,2])
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()
```

```python
line_model = keras.Sequential()
```

```python
line_model.add(keras.layers.Dense(2, activation='sigmoid', input_shape=(2,)))
line_model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
optimizer = keras.optimizers.SGD(lr=0.1)
```

```python
line_model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
line_model.fit(x_data, y_data, batch_size=2, epochs=200)
```

```python
predict_y_data = line_model.predict(x_data)
```

```python
np.round(predict_y_data)
```
