## sine 그래프

```python
xdata = np.arange(1, 6, 0.1)
ydata = np.sin(xdata)
sin_e = plt.plot(xdata, ydata, label='Sin e')
plt.savefig('tane.png')
plt.legend()
plt.show()
```

## cose 그래프

```python
xdata = np.arange(1, 6 ,0.1)
ydata = np.cos(xdata)
cos_e = plt.plot(xdata, ydata, label='Cos e')
plt.savefig('Cose.png')
plt.legend()
plt.show()
```

```python
xdata = np.arange(1, 6, 0.1)
#plt.plot(xdata, np.sin(xdata), label = 'sin e', linestyle='-') # default : '-'
#plt.plot(xdata, np.sin(xdata), label = 'sin e', linestyle='--')
plt.plot(xdata, np.sin(xdata), label = 'sin e', linestyle=':')
plt.plot(xdata, np.cos(xdata), label = 'cos e', linestyle='dashdot')
plt.legend()
plt.show()
```

## tane 그래프

```python
xdata = np.arange(-1, 1, 0.01)
ydata = np.tan(xdata)
plt.plot(xdata, ydata, label='tan e')
plt.plot([0], [0], 'o')
plt.plot([-1], [-1], 'o')
plt.plot([1], [1], 'o')
plt.savefig('tane.png')
plt.legend()
plt.show()
```

```python
xdata = np.arange(1, 6, 0.1)
ydata = np.sin(xdata)/np.cos(xdata)
plt.plot(xdata, ydata, label='tan e')
plt.savefig('tane.png')
plt.legend()
plt.show() # 컴퓨터는 무한대 값을 표현할 수 없어서 그래프가 깨짐.
```

```python
plt.plot([3], [3], 'x') # 점을 찍을 때는 'o' ('x'도 가능)
plt.show()
```

## circle 그래프

### plt.pie([데이터1, 데이터2, ...], labels=[데이터개수만큼], colors=[])

```python
m = 100 # male
f = 150 # female

m_per = m / (m+f) # male 비율
f_per = f / (m+f) # female 비율

color1 = ['skyblue', 'pink']
label1 = ['male', 'female']

plt.figure(figsize = (10, 10)) # 원의 크기 figsize=(최대크기, 원크기)
patches,texts = plt.pie([m_per, f_per], labels=label1, colors=color1)
```

### 부산 -> 대구, 대전, 인천

```python
b_df = pd.read_excel("101_DT_1B26003_A01_20200704153647.xlsx", header=1)
b_df.head(3)
```

```python
mask1 = (b_df['전출지별'] == '부산광역시') & (b_df['전입지별'] != '부산광역시') & (b_df['항목'] == '이동자수')
```

```python
b_df1 = b_df[mask1]
b_df1.head(3)
```

```python
b_df2 = b_df1[list(b_df1)[:3] + sorted(list(b_df1)[4:])].fillna(0)
b_df2.head(3)
```

```python
b_df2.set_index('전입지별', inplace=True)
b_df2.head(3)
```

```python
b_df2.drop(['전출지별', '항목'], axis=1, inplace=True)
```

```python
b_df2
```

```python
year = list(map(lambda x:str(x) + ' 년', range(2000, 2020)))
year
```

```python
b_df3 = b_df2.loc[['대구광역시', '대전광역시', '인천광역시'], year]
```

```python
b_df3
```

```python
b_df4 = b_df3.transpose()
b_df4.plot(kind='bar', color=['red', 'green', 'blue'])
```
