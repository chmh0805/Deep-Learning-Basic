## 로또 번호 추천 프로그램

```python
import random

num_list = [x for x in range(1, 46)] # 로또 번호 1 ~ 45 [1, 2, 3, ... , 44, 45]
num_count = []
result = [] # 추천 숫자 6개가 담길 그릇
num = 1
cnt = [0*x for x in range(46)]

while num < 20:
    print("현재 단계 : {}".format(num))
    for step in range(10001):
        num_count.append(random.randint(1, 45))
        if step % 500 == 0:
            print("\n현재 step == {}\n".format(step))
    print("============================================\n")
    num += 1

for i in num_count:
    cnt[i] += 1

while len(result) < 6:
    result.append(cnt.index(max(cnt)))
    cnt[cnt.index(max(cnt))] = 0

print("================")
print("추천 번호는\n")
for num in sorted(result):
    print("{} ".format(num))
print("\n입니다.\n")
print("================")
```
