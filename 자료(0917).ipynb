### 그래프 라벨 적용

```python
plt.plot(ar_df.index,ar_df['pow'],label='x^2') #label 변수를 통해서 그래프명 지정
plt.plot(np.arange(1,100),np.arange(1,100)*2, label='2x')
plt.legend() #legend() 함수를 통해서 라벨 적용
plt.savefig('graph1.png')
plt.show()
```

### sine, cosine, tangent 그래프

```python
num_ar=np.arange(1,6,0.1)
```

```python
x=num_ar
y=np.sin(x)
y
```

```python
plt.plot(x,y,label='sine')
plt.legend()
plt.savefig('sine.png')
plt.show()
```

```python
plt.plot(num_ar,np.cos(num_ar),label='cosine')
plt.legend()
plt.savefig('cosine.png')
plt.show()
```

```python
#plt.plot(num_ar,np.sin(num_ar),label='sine',linestyle='--')
plt.plot(num_ar,np.sin(num_ar),label='sine',linestyle='-.')
plt.plot(num_ar,np.cos(num_ar),label='cosine',linestyle=':')
plt.legend()
plt.show()
```

```python
num_ar1=np.arange(-1.5,1.5,0.1)
plt.plot(num_ar1,np.tan(num_ar1),label='tan')
plt.show()
```

```python
plt.plot(num_ar,np.sin(num_ar),label='sine')
plt.plot(num_ar,np.cos(num_ar),label='cosine')
plt.plot(num_ar,np.tan(num_ar),label='tan')
plt.legend()
plt.show()
```

```python
a_df=pd.read_excel('101_DT_1B26003_A01_20200704153647.xlsx',header=1)
```

```python
mask=(a_df['전출지별']=='서울특별시') & (a_df['전입지별'] != '서울특별시') & (a_df['항목']=='이동자수')
```

```python
a_df1=a_df[mask]
```

```python
a_df1
```

```python
a_df1.columns
```

```python
list(a_df1.columns)
```

```python
a_df2=a_df1[list(a_df1)[:3]+sorted(list(a_df1)[4:])].fillna(0)
```

```python
a_df2.head(3)
```

```python
a_df3=a_df2.drop(axis=0,columns=['전출지별','항목'])
```

```python
a_df3
```

```python
a_df3.set_index('전입지별',inplace=True)
```

```python
a_df3
```

```python
#서울->경기도 전입한 인구수
a_df3_n=a_df3.loc['경기도']
```

```python
a_df3_n.index
```

```python
a_df3_n.values
```

```python
#font 설정
import matplotlib.font_manager as fm
path='c:\\windows\\fonts\\malgunsl.ttf'
font_name=fm.FontProperties(fname=path,size=20).get_name()
plt.rc('font',family=font_name)

plt.plot(a_df3_n.index,a_df3_n.values)
plt.title('서울=>경기')
plt.xlabel('시기')
plt.ylabel('인구수')
plt.show()
```

```python
plt.figure(figsize=(20,5)) # 그래프 크기
plt.plot(a_df3_n.index,a_df3_n.values,label='서울=>경기')
plt.xticks(rotation='vertical') # x축 세로로 설정
plt.title('서울=>경기')
plt.xlabel('시기')
plt.ylabel('인구수')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.style.use('bmh') # 그래프 바탕색 설정
plt.plot(a_df3_n.index,a_df3_n.values,label='서울=>경기')
plt.xticks(rotation='vertical')
plt.title('서울=>경기')
plt.xlabel('시기')
plt.ylabel('인구수')
plt.show()
```
