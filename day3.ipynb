```python
from tensorflow.compat.v1 import ConfigProto, InteractiveSession
config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.4
session = InteractiveSession(config = config)
```

```python
data = [
    [0.72,0.82,-1],
    [0.91,-0.69,-1],
    [0.03,0.93,-1],
    [0.12,0.25,-1],
    [0.96,0.47,-1],
    [0.8,-0.75,-1],
    [0.46,0.98,-1],
    [0.66,0.24,-1],
    [0.72,-0.15,-1],
    [0.35,0.01,-1],
    [-0.11,0.1,1],
    [0.31,-0.96,1],
    [0.0,-0.26,1],
    [-0.43,-0.65,1],
    [0.57,-0.97,1],
    [-0.72,-0.64,1],
    [-0.25,-0.43,1],
    [-0.12,-0.9,1],
    [-0.58,0.62,1],
    [-0.77,-0.76,1]
]
data=np.array(data)
```

```python
x_data=data[:,:2]
y_data=(data[:,2:]+1)/2
```

```python
plt.scatter(x_data[:,0], x_data[:,1], c=y_data)
plt.show()
```

```python
model_1 = keras.Sequential()
```

```python
model_1.add(keras.layers.Dense(1, activation='sigmoid', input_shape=(2,)))
```

```python
optimizer = keras.optimizers.SGD(lr=0.1)
```

```python
model_1.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_1.fit(x_data, y_data, batch_size=5, epochs=2000)
```

```python
model_1.predict(x_data)
```

```python
model_1.layers[0].get_weights()
```

```python
weight = model_1.layers[0].get_weights()[0]
bias = model_1.layers[0].get_weights()[1]
x = np.linspace(np.amin(x_data[:,:1]), np.amax(x_data[:,:1]))
x
```

```python
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
a=-weight[0]/weight[1]
b=-(weight[0]/weight[1])/(weight[1]/bias[0])
plt.plot(x,[a*i +b for i in x],color='green')
plt.show()
```

# MNIST

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as  plt
from tensorflow.keras.datasets import mnist
```

##  

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train[0]
```

```python
plt.imshow(x_train[0], 'gray')
plt.show()
```

##  

```python
x_train.shape
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```python
x_train.shape
```

###  

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
y_train.shape
```

### y one-hot-encoding

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
y_train[0]
```

```python
y_train.shape
```

## Mnist  

```python
mnist_model = keras.Sequential()
```

```python
mnist_model.add(keras.layers.Dense(50, activation='relu', input_shape=(784,)))
mnist_model.add(keras.layers.Dense(100, activation='relu'))
mnist_model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
mnist_model.summary()
```

```python
mnist_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
mnist_model.fit(x_train, y_train, batch_size=500, epochs=20)
```

##  

```python
y_predict = np.round(mnist_model.predict(x_test))
y_predict[0]
```

```python
y_predict = y_predict.argmax(axis=1)
y_test = y_test.argmax(axis=1)
```

```python
y_predict[0]
```

```python
y_test[0]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_predict)
```

```python
miss = np.where(y_test != y_predict) #   .
```
