## AndGate 구현

```python
from tensorflow import keras
import numpy as np
```

```python
from tensorflow.compat.v1 import ConfigProto,InteractiveSession
config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction=0.4
session = InteractiveSession(config=config)
```

### 데이터 수집

```python
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]
y_and_data = [[0],
             [0],
             [0],
             [1]]
```

### 데이터 가공

```python
x_data = np.array(x_data)
y_and_data = np.array(y_and_data)
```

```python
x_data.shape
```

```python
y_and_data.shape
```

### 모델 선정

```python
model_and = keras.Sequential()
```

```python
model_and.add(keras.layers.Dense(2, activation='sigmoid', input_shape=(2,)))
model_and.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
optimizer = keras.optimizers.SGD(lr=0.1)
```

```python
model_and.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

### 모델학습

```python
model_and.fit(x_data, y_and_data, batch_size=4, epochs=3000)
```

### 모델 평가

```python
model_and.predict(x_data)
```

```python
np.round(model_and.predict(x_data))
```

## ORGate 구현

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
x_data = [[0, 0],
         [0, 1],
         [1, 0],
         [1, 1]]
y_or_data = [[0],
            [1],
            [1],
            [0]]
```

```python
x_data = np.array(x_data)
y_or_data = np.array(y_or_data)
```

```python
model_or = keras.Sequential()
```

```python
model_or.add(keras.layers.Dense(2, activation='sigmoid', input_shape=(2,)))#입력층
model_or.add(keras.layers.Dense(1, activation='sigmoid'))#출력층
```

```python
optimizer = keras.optimizers.SGD(lr=0.1) # SGD 기법, lr = 0.1
```

```python
model_or.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_or.fit(x_data, y_or_data, batch_size=4, epochs=3000)
```

```python
model_or.predict(x_data)
```

```python
np.round(model_or.predict(x_data))
```

### XORGate 구현

```python
x_data = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]
y_xor_data = [
    [0],
    [1],
    [1],
    [0]
]
```

```python
x_data = np.array(x_data)
y_xor_data = np.array(y_xor_data)
```

```python
model_xor = keras.Sequential()
```

```python
model_xor.add(keras.layers.Dense(2, activation='sigmoid', input_shape=(2,)))
model_xor.add(keras.layers.Dense(32, activation='relu', input_shape=(2,)))
model_xor.add(keras.layers.Dense(1, activation='relu'))
```

```python
optimizer = keras.optimizers.SGD(lr=0.1)
model_xor.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['accuracy'])
```

```python
model_xor.fit(x_data, y_xor_data, batch_size=4, epochs=2000)
```
